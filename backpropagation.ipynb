## Keterangan
$\begin{align}
x&&&\text{Vektor masukan}\\
&&&\hspace{0.5cm}x=(x_1,\dots,x_i,\dots,x_n)\\
t&&&\text{Target vektor keluaran}\\
&&&\hspace{0.5cm}t=(t_1,\dots,t_k,\dots,t_m)\\
\sigma_k&&&\text{Bagian dari penyesuaian bobot koreksi kesalahan untuk $W_{jk}$ yang disebabkan oleh kesalahan pada unit keluaran $Y_k$.}\\
&&&\text{Informasi tentang kesalahan pada unit $Y_k$ yang disebarkan kembali ke unit tersembunyi yang dimasukkan ke dalam unit $Y_k$}\\
\delta_j&&&\text{Bagian dari penyesuaian bobot koreksi kesalahan untuk $V_{ij}$ yang disebabkan oleh propagasi balik informasi kesalahan}\\
&&&\text{dari lapisan keluaran ke unit tersembunyi $z$}\\
\alpha&&&\text{Laju pembelajaran}\\
X_i&&&\text{Unit masukan $i$}\\
&&&\text{Untuk setiap unit masukan, sinyal masukan dan sinyal keluaran sama, dinamakan $x_i$}\\
v_{0j}&&&\text{Bias dari unit tersembunyi $j$}\\
Z_j&&&\text{Unit tersembunyi $j$}\\
&&&\text{Masukan bersih dari $Z_j$ dilambangkan $z\_in_j$}\\
&&&\hspace{0.5cm}z\_in_{j}=v_{0j}+\sum\limits_{i}x_iv_{ij}\\
&&&\text{Sinyal keluaran dari $Z_j$ dilambangkan $z_j$}\\
&&&\hspace{0.5cm}z_{j}=f(z\_in_j)\\
w_{0k}&&&\text{Bias dari unit keluaran $k$}\\
Y_k&&&\text{Unit keluaran $k$}\\
&&&\text{Masukan bersih dari $Y_k$ dilambangkan $y\_in_k$}\\
&&&\hspace{0.5cm}y\_in_{k}=w_{0k}+\sum\limits_{j}z_jw_{jk}\\
&&&\text{Sinyal keluaran dari $Y_k$ dilambangkan $y_j$}\\
&&&\hspace{0.5cm}y_{k}=f(y\_in_k)
\end{align}$

## Algoritma
---

_**Step 0**_ Inisialisasi bobot <br>
_**Step 1**_ Ketika kondisi berhenti bernilai *false*, kerjakan _Step 2-9_ <br>
$\hspace{1.1cm}$_**Step 2**_ Untuk setiap pasangan latih, kerjakan _Step 3-8_ <br>
$\hspace{2.3cm}$_Feedforward_ <br>
$\hspace{2.3cm}$_**Step 3**_ Setiap unit input ($X_i,i=1,\dots,n$) menerima sinyal masukan $x_i$ dan <br>
$\hspace{3.5cm}$ menyebarkan sinyal ini ke semua unit di lapisan diatasnya (unit tersembunyi) <br>
$\hspace{2.3cm}$_**Step 4**_ Setiap unit tersembunyi ($Z_j,j=1,\dots,p$) menjumlahkan bobot-bobot dari sinyal input,<br>
$\hspace{4cm}$ $z\_in_{j}=v_{0j}+\sum\limits_{i=1}^nx_iv_{ij}$, <br>
$\hspace{3.5cm}$ gunakan fungsi aktivasi untuk menghitung sinyal keluaran, <br>
$\hspace{4cm}$ $z_{j}=f(z\_in_j)$, <br>
$\hspace{3.5cm}$ dan sinyal ini dikirim ke semua unit di lapisan diatasnya (sinyal keluaran)<br>
$\hspace{2.3cm}$_**Step 5**_ Setiap unit keluaran ($Y_k,k=1,\dots,m$) menjumlahkan bobot-bobot dari sinyal masukan,<br>
$\hspace{4cm}$ $y\_in_{k}=w_{0k}+\sum\limits_{j=1}^pz_jw_{jk}$, <br>
$\hspace{3.5cm}$ gunakan fungsi aktivasi untuk menghitung sinyal keluaran, <br>
$\hspace{4cm}$ $y_{k}=f(y\_in_k)$ <br>
$\hspace{2.3cm}$_Backpropagation of error_ <br>
$\hspace{2.3cm}$_**Step 6**_ Setiap unit keluaran ($Y_k,k=1,\dots,m$) menerima pola target <br>
$\hspace{3.5cm}$ yang berkorelasi dengan pola latih, lalu dihitung nilai errornya <br>
$\hspace{4cm}$ $\delta_k=(t_k-y_k)f'(y\_in_k)$, <br>
$\hspace{3.5cm}$ hitung koreksi bobot (digunakan untuk memperbarui $w_{jk}$ nanti) <br>
$\hspace{4cm}$ $\triangle w_{jk}=\alpha \delta_kz_j$, <br>
$\hspace{3.5cm}$ hitung koreksi bias (digunakan untuk memperbarui $w_{0k}$ nanti) <br>
$\hspace{4cm}$ $\triangle w_{0k}=\alpha \delta_k$, <br>
$\hspace{3.5cm}$ dan $\delta_k$ dikirim ke semua unit di lapisan bawahnya<br>
$\hspace{2.3cm}$_**Step 7**_ Setiap unit tersembunyi ($Z_j,j=1,\dots,p$) menjumlahkan delta masukan (dari unit lapisan diatasnya)<br>
$\hspace{4cm}$ $\delta\_in_j=\sum\limits_{k=1}^m\delta_kw_{jk}$, <br>
$\hspace{3.5cm}$ hasilnya dikalikan dengan turunan fungsi aktivasi yang digunakan untuk menghitung nilai error, <br>
$\hspace{4cm}$ $\delta_j=\delta\_in_jf'(z\_in_j)$, <br>
$\hspace{3.5cm}$ hitung koreksi bobot (digunakan untuk memperbarui $v_{ij}$ nanti) <br>
$\hspace{4cm}$ $\triangle v_{ij}=\alpha \delta_jx_i$, <br>
$\hspace{3.5cm}$ hitung koreksi bobot (digunakan untuk memperbarui $v_{0j}$ nanti) <br>
$\hspace{4cm}$ $\triangle v_{0j}=\alpha \delta_j$, <br>
$\hspace{2.3cm}$ Perbarui bobot dan bias <br>
$\hspace{2.3cm}$_**Step 8**_ Setiap unit keluaran ($Y_k,k=1,\dots,m$) memperbarui bobot dan biasnya ($j=0,\dots,p$)<br>
$\hspace{4cm}$ $w_{jk}(baru)=w_{jk}(lama)+\triangle w_{jk}$<br>
$\hspace{3.5cm}$ Setiap unit tersembunyi ($Z_j,j=1,\dots,p$) memperbarui bobot dan biasnya ($i=0,\dots,n$) <br>
$\hspace{4cm}$ $v_{ij}(baru)=v_{ij}(lama)+\triangle v_{ij}$<br>
$\hspace{1.1cm}$_**Step 9**_ Tes kondisi berhenti

---

## Implementasi

```python
import numpy as np
```

```python
class Backpropagation():
    def __init__(self,neuron=100, learning_rate=1, max_epoch=100):
        self.v = None
        self.w = None
        self.neuron = neuron
        self.learning_rate = learning_rate
        self.max_epoch = max_epoch
    def fit(self, x, t):
        if t.ndim==1:
            t = np.expand_dims(t,axis=1)
        self.v, self.w = trainBackpropagation(x, t, self.neuron, self.learning_rate, self.max_epoch)
    def predict(self, x):
        return testBackpropagation(x, self.v, self.w)
```

```python
def trainBackpropagation(x, t, neuron=100, learning_rate=1, max_epoch=100):
    # Step 0
    v = np.random.rand(x.shape[1]+1, neuron)
    w = np.random.rand(v.shape[1]+1, t.shape[1])
    
    delta_v = np.zeros(v.shape)
    delta_w = np.zeros(w.shape)
    
    epoch = 1
    stop = False
    
    # Step 1
    while not stop and epoch<max_epoch:
        stop = True
        # Step 2
        for i in range(t.shape[0]):
            # Step 3 adalah x
            
            # Step 4
            z_in = v[0,:] + np.dot(x[i,:], v[1:,:])
            z_unit = f(z_in)
            
            # Step 5
            y_in = w[0,:] + np.dot(z_unit, w[1:,:])
            y_unit = f(y_in)
            
            # Step 6
            sigma_y = (t[i,:]-y_unit)*f(y_in, deriv=True)
            
            delta_w[0,:] = learning_rate*sigma_y
            delta_w[1:,:] = learning_rate*sigma_y*np.hsplit(z_unit, len(z_unit))
            
            # Step 7
            sigma_in_z = sum((sigma_y*w[1:,:]).T)
            sigma_z = sigma_in_z*f(z_in, deriv=True)
            
            delta_v[0,:] = learning_rate*sigma_z
            delta_v[1:,:] = learning_rate*sigma_z*np.hsplit(x[i,:], len(x[i,:]))
            
            # Step 8
            w += delta_w
            v += delta_v
        
        # Step 9
        if not np.array_equal(delta_w,np.zeros(delta_w.shape)) or not np.array_equal(delta_v,np.zeros(delta_v.shape)):
            stop = False
        rmse = RMSE(t, testBackpropagation(x,v,w))
        if rmse < 0.1:
            stop = True
        
        epoch+=1
    print("Max Epoch:",epoch,"\t","Max RMSE:",rmse)
    return v,w
```

```python
def testBackpropagation(x, v, w):
    z_in = v[0,:] + np.dot(x, v[1:,:])
    z_unit = f(z_in)
    y_in = w[0,:] + np.dot(z_unit, w[1:,:])
    y_unit = f(y_in)
    return y_unit
```

```python
# Fungsi aktivasi sigmoid
def f(x, deriv=False):
    if not deriv:
        return 1/(1+np.exp(-x))
    return f(x)*(1-f(x))
```

```python
# Fungsi performasi dengan meninjau rmse paling besar dari setiap target
def RMSE(act, pred):
    n = act.shape[0]
    mse = [(pred[i,:]-act[i,:])**2/n for i in range(n)]
    rmse = np.sqrt(sum(mse))
    return max(rmse)
```
